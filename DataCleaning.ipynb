```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_rows', 125)
```

```python
application_data = pd.read_csv('Data/application_data.csv')
prev_appln_data = pd.read_csv('Data/previous_application.csv')
```

```python
def missing_data_finder(df, missing_value):
    missing_data = pd.DataFrame(df.isna().sum()/len(df)*100)
    missing_data.rename(columns={0:"Missing%"}, inplace=True)
    cols_to_include = missing_data[missing_data['Missing%']<=missing_value]
    cols_to_include = cols_to_include.index.values.tolist()
    print("Dataset compressed from {} columns to {} columns".format(df.shape[1], len(cols_to_include)))
    return cols_to_include
```

```python
application_cols_to_include = missing_data_finder(application_data, 45)
```

```python
prev_appln_cols_to_include = missing_data_finder(prev_appln_data, 45)
```

```python
appln_data = pd.read_csv('Data/application_data.csv', usecols=application_cols_to_include)
appln_data.head()
```

```python
prev_appln_data = pd.read_csv('Data/previous_application.csv', usecols=prev_appln_cols_to_include)
prev_appln_data.head()
```

```python
# To remove the rest of the missing values, we will use listwise deletion
```

```python
appln_data.dropna(axis=1,inplace=True)
prev_appln_data.dropna(axis=1,inplace=True)
```

```python
# Create new dataframes for the cleaned datasets.

appln_data.to_csv("appln_data.csv")
prev_appln_data.to_csv("prev_appln_data.csv")
```
